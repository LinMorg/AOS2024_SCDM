## Map of Antarctica and surrounding Southern ocean

```python
#import packages
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

#set extent and plot
extent= [-180, 180, -90, -60]
ax= plt.axes(projection=ccrs.SouthPolarStereo())
ax.set_extent(extent, ccrs.PlateCarree())
ax.coastlines(resolution='50m')
ax.gridlines(draw_labels=True)
ax.stock_img()
ax.set_title("Antartica")
```

Map of Antarctica and the Southern Ocean, the projection chosen is the SouthPolarStereo. This is becuase the required area was Antarctic, thus a projection with the southern ocean as its focus would be best suited. 

## The south Atlantic

```python
#Import packages
import matplotlib.pyplot as plt
import cartopy.crs as ccrs 
import cartopy.feature as cfeature

#set locations    
walvis_bay= dict(lon= 14.5058, lat= -22.9576)
cape_town= dict(lon= 18.4241, lat= -33.9249)
rio_de_janerio= dict(lon= -43.1729, lat= -22.9068)
montevideo= dict(lon= -56.1645, lat= -34.9011)
lons= [walvis_bay['lon'], cape_town['lon'], rio_de_janerio['lon'], montevideo['lon']]
lats= [walvis_bay['lat'], cape_town['lat'], rio_de_janerio['lat'], montevideo['lat']]
extent= [-800, 20, -60, -20]

#plot locations and features
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent(extent)
ax.coastlines(resolution='50m')
ax.scatter(walvis_bay['lon'], walvis_bay['lat'], color='blue')
ax.scatter(cape_town['lon'], cape_town['lat'], color='red')
ax.scatter(rio_de_janerio['lon'], rio_de_janerio['lat'], color='green')
ax.scatter(montevideo['lon'], montevideo['lat'], color= 'purple')
ax.stock_img()
ax.add_feature(cfeature.RIVERS)


#labels
ax.text(10.5058, -26.9576, "Walvis Bay", fontsize=10)
ax.text(13.4241, -36.9249, 'Cape Town', fontsize=10)
ax.text(-40.1729, -22.9068, "Rio de Janerio", fontsize=10)
ax.text(-54.1645, -36.9011, 'Montevideo', fontsize=10)
gl= ax.gridlines(draw_labels=True, linewidth=1, color='gray', linestyle='-')
gl.right_labels = False
gl.top_labels= False
ax.set_title('South Atlantic')
```

Map of the South Atlantic, with the cities Walvis Bay, Cape Town, Rio de Janerio and Montevideo indicated. Projection chosen was PlateCarree, as  this projection is best suited for regions spanning over long longitudes. 

## False Bay in different resolutions

```python
#Import packages
import matplotlib.pyplot as plt
import cartopy.crs as ccrs 
import cartopy.feature as cfeature

#set resolutions
resolutions= ['110m', '50m', '10m']

#set titles
titles= ['Coarse Resolution 110m', 'Intermediate resolution 50m', 'Full resolution 10m']

#plot map
fig, axs= plt.subplots(1, 3, figsize= (15, 5), subplot_kw=dict(projection=ccrs.Orthographic()))
titles= ['Coarse Resolution 110m', 'Intermediate resolution 50m', 'Full resolution 10m']
for ax, resolution, title in zip(axs, resolutions, titles):
    ax.set_extent([18, 19.5, -34.5, -33.5], crs=ccrs.PlateCarree())
    ax.add_feature(cfeature.LAND, color='lightgray')
    ax.add_feature(cfeature.OCEAN, color='white')
    ax.add_feature(cfeature.COASTLINE.with_scale(resolution))
    gl= ax.gridlines(draw_labels=True, linewidth=1, color='gray', linestyle='-')
    gl.right_labels = False
    gl.top_labels= False
    ax.set_title(title)
```

False bay is resolutions ranging from coarse, 110m, to full, 10m. Projection chosen was Orthographic, as this projection is the most true with regards to reality. The grey land is to show how the resolution losses its detail the more coarse it is.
